## Abstract

This project examines how listing features in San Diego Airbnb listings are associated with expectation disconfirmation between host-written descriptions and guest-reported experiences. We measured this using an Expectation Gap metric, defined as the difference between the VADER sentiment score of the listing description and the mean VADER sentiment score of its reviews, where a gap near 0 indicates the description sentiment matched review sentiment. 

We found that the description length is the strongest positive predictor of a larger gap, while super host status is associated with a smaller, more favorable gap. Additionally, listings with smaller absolute gaps average about 11 more booked nights per nights per year than listings with larger expectation gaps. Overall, these results suggest that hosts benefit from writing more calibrated, grounded descriptions, since setting accurate expectations is linked to both better alignment with guest experiences and higher booking demand.

**For San Diego Airbnb listings (September 2025 snapshot), how do listing-level factors (superhost status, nightly price, neighborhood, room type, number of amenities, and description language characteristics) impact expectation disconfirmation between host-written listing descriptions and guest-reported experiences, as measured by our computed Expectation Gap metric?**

According to Oliver (1977), **disconfirmation** is defined as the extent to which a consumer's initial expectations about a product's performance are met. It operates on a continuum ranging from unfavorable to favorable. Specifically, negative disconfirmation occurs when a product performs worse than expected, confirmation happens when it performs exactly as expected, and positive disconfirmation occurs when it performs better than anticipated.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

For this project, we measure disconfirmation using a custom metric called the **Expectation Gap**, defined as:

> **Expectation Gap = VADER(description) − mean(VADER(reviews))**

where VADER produces a compound sentiment score from −1 (most negative) to +1 (most positive). A gap near 0 indicates the description sentiment matched review sentiment (confirmation); a positive gap means the host oversold (negative disconfirmation); a negative gap means the host undersold, where guests were pleasantly surprised (positive disconfirmation).

This leads to three specific sub-questions:
- Which listing features best explain why some listings have a smaller Expectation Gap (absolute gap closer to 0)?
- Which features are the strongest predictors of gap size?
- How does the Expectation Gap relate to booking demand (estimated nights booked per year)?

## Background and Prior Work

Online peer-to-peer (P2P) marketplaces rely heavily on trust between users, and platforms like Airbnb are no exception. Because guests cannot physically visit a property before booking, they depend largely on host-provided listing descriptions, photos, and amenities to form expectations about their stay. Expectation–confirmation theory (ECT), developed by Oliver and widely applied in hospitality contexts, models satisfaction as this comparison between prior expectations and perceived performance; negative disconfirmation (worse than expected) leads to dissatisfaction, while confirmation or positive disconfirmation leads to satisfaction.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) When guests' experiences deviate significantly from their expectations, the resulting expectation gap can lead to lower ratings, negative reviews, reduced future demand, and erosion of platform trust.<br><br>
​
Text‑mining work on Airbnb reviews shows what guests actually care about. Cheng and Jin analyze a large corpus of Airbnb review comments using topic modeling and sentiment analysis, finding that experiences are dominated by location, amenities, and host interaction, with a strong overall positivity bias in reviews.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Other work confirms that Airbnb reviews are overwhelmingly positive and that criticism is often muted, which makes star ratings alone a poor indicator of underlying problems.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)<br>

Expectation disconfirmation has been explicitly studied in the Airbnb context: Zhang et al. show that mismatches between expectations and experience, and how consistent reviews are, meaningfully affect guest attitudes and intention to recommend listings.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) Most directly related to this project, Cavique et al. measure semantic alignment between Airbnb host descriptions and guest reviews using word embeddings and clustering. They introduce a concept coverage ratio (how many review “concepts” are also present in the description) and find that higher coverage is generally associated with more positive review sentiment.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)<br><br>
​
Building on this work, the proposed project defines a new expectation gap score that combines global semantic similarity between descriptions and reviews with host–guest sentiment differences (capturing overpromising vs. underpromising). It then uses supervised models to predict this score from pre-booking information (descriptions, amenities, host and listing features). In a diverse tourism market like San Diego, where listings range from luxury beachfront villas in La Jolla to urban studios in Little Italy, the factors contributing to expectation gaps are multifaceted. By quantifying which listing characteristics predict smaller expectation gaps, this project aims to provide actionable insights for hosts to improve transparency and for platforms to better calibrate guest expectations.

## Hypothesis

We predict that listings with superhost status, more amenities, and grounded description language will have a smaller Expectation Gap (absolute gap closer to 0). Listings with smaller gaps will also have higher estimated booking frequency, measured by `estimated_occupancy_l365d` (estimated nights booked in the last 365 days).

- **Superhost status:** Superhosts must maintain ≥4.8 rating, ≥90% response rate, and ≥10 stays/year to keep their badge, so they are more likely to write honest, calibrated descriptions.
- **Amenity count:** More amenities give hosts concrete features to describe rather than relying on vague language, which should help set expectations that the listing can actually deliver on.
- **Description language:** A longer or more positively-worded description signals a higher risk of overpromising; we operationalize "grounded" as a VADER description score that does not exceed the listing's mean review VADER score by more than 0.2 units.

## Results

### Exploratory Data Analysis

Before running statistical tests, we first explore the data visually to understand how the Expectation Gap is distributed, which groups of listings tend to have larger or smaller gaps, and which variables seem related to it. The four sections below each look at a different angle of the data.


#### Plot 1: What Does the Expectation Gap Look Like Overall and Does Superhost Status Matter?


```python
# Plot 1: Gap distribution + Gap by superhost status
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Distribution
axes[0].hist(listings_gap['expectation_gap'].dropna(), bins=50, edgecolor='black', alpha=0.7, color='steelblue')
axes[0].axvline(0, color='red', linestyle='--', linewidth=1.5, label='No gap (0)')
axes[0].set(xlabel='Expectation Gap', ylabel='Count',
            title='Distribution of Expectation Gap\n(VADER description − mean VADER reviews)')
axes[0].legend()

# Superhost boxplot
sh_data = listings_gap[listings_gap['host_is_superhost'].notna()]
data = [sh_data[sh_data['host_is_superhost'] == True]['expectation_gap'].dropna(),
        sh_data[sh_data['host_is_superhost'] == False]['expectation_gap'].dropna()]
bp = axes[1].boxplot(data, labels=['Superhost', 'Non-Superhost'], patch_artist=True,
                     medianprops=dict(color='red', linewidth=2))
bp['boxes'][0].set_facecolor('lightblue')
bp['boxes'][1].set_facecolor('lightyellow')
axes[1].axhline(0, color='gray', linestyle='--', alpha=0.5, label='No gap (0)')
axes[1].set(ylabel='Expectation Gap', title='Expectation Gap by Superhost Status')
axes[1].legend()

plt.tight_layout()
plt.savefig('results/plot1_distribution_superhost.png', dpi=150, bbox_inches='tight')
plt.show()
```

The left plot shows the full distribution of Expectation Gap scores across all 7,879 listings. The right plot compares the gap between superhost and non-superhost listings using boxplots, where the box covers the middle 50% of values and the line inside is the median.

The histogram shows that most listings have a gap close to zero, meaning the host's description and guest reviews have similar sentiment. The distribution is slightly left-skewed, most listings actually have a small **positive** gap (the host's description is a bit more positive than the reviews), but a smaller group of listings has very large negative gaps, where guests were dramatically more positive than the description suggested.

The red dashed line marks zero: the point where description perfectly matched reviews. Most listings sit just to the right of it, meaning hosts tend to modestly oversell. The long left tail shows a subset of listings where guests were far more positive than the description implied.

The boxplots show that both superhosts and non-superhosts have similar typical gaps (medians near zero). However, non-superhosts have more extreme positive-gap outliers they are more likely to write descriptions that end up overselling the experience. Superhosts show a slightly tighter, more consistent distribution.


```python
# Plot 2: Gap by room type + top 10 neighborhoods
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Room type boxplot
room_types = listings_gap['room_type'].value_counts()
valid_rooms = room_types[room_types > 20].index
room_data = [listings_gap[listings_gap['room_type'] == rt]['expectation_gap'].dropna()
             for rt in valid_rooms]
bp2 = axes[0].boxplot(room_data, labels=valid_rooms, patch_artist=True,
                      medianprops=dict(color='red', linewidth=2))
colors = ['lightblue', 'lightgreen', 'lightyellow', 'lightsalmon']
for patch, color in zip(bp2['boxes'], colors):
    patch.set_facecolor(color)
axes[0].axhline(0, color='gray', linestyle='--', alpha=0.5)
axes[0].set(ylabel='Expectation Gap', title='Expectation Gap by Room Type')
axes[0].tick_params(axis='x', rotation=15)

# Top 10 neighborhoods bar chart
top10 = listings_gap['neighbourhood_cleansed'].value_counts().head(10).index
gap_by_hood = (listings_gap[listings_gap['neighbourhood_cleansed'].isin(top10)]
               .groupby('neighbourhood_cleansed')['expectation_gap'].mean()
               .sort_values())
colors_bar = ['tomato' if v > 0 else 'steelblue' for v in gap_by_hood.values]
axes[1].barh(gap_by_hood.index, gap_by_hood.values, color=colors_bar)
axes[1].axvline(0, color='black', linestyle='-', linewidth=1)
axes[1].set(xlabel='Mean Expectation Gap', title='Mean Gap by Neighborhood (Top 10 by listing count)')

plt.tight_layout()
plt.savefig('results/plot2_roomtype_neighborhood.png', dpi=150, bbox_inches='tight')
plt.show()
```

The left plot shows the spread of Expectation Gap for each room type. Each box covers the middle 50% of listings in that category, with dots representing extreme outliers.

**Room type:** Entire home/apartment listings, by far the most common room type (~88% of listings), have the widest spread and the most extreme outliers in both directions. This is likely because entire-home listings in San Diego vary enormously, from beachfront houses to urban studios. Private room listings cluster closer to zero, suggesting more consistent alignment between descriptions and guest experiences. Shared rooms have a small sample size but lean slightly positive, meaning hosts in that category tend to oversell slightly.

**Neighborhoods:** The right bar chart shows the average Expectation Gap for San Diego's 10 most-listed neighborhoods. Bars to the left of zero mean guests were more positive than the description on average, and nearly every top neighborhood shows this pattern. Park West is the exception with a slight positive gap, meaning descriptions there tend to set expectations that guests find harder to meet, possibly because the neighborhood's urban density and parking limitations are not always made clear upfront.


```python
# Plot 3: Gap vs price (binned) + gap vs amenity count
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Price bins
listings_gap['price_bin'] = pd.cut(listings_gap['price'].clip(upper=1000), bins=8)
price_means = listings_gap.groupby('price_bin', observed=True)['expectation_gap'].mean()
axes[0].plot(range(len(price_means)), price_means.values, 'o-', color='coral', linewidth=2, markersize=6)
axes[0].set_xticks(range(len(price_means)))
axes[0].set_xticklabels([f'${int(iv.left)}–{int(iv.right)}' for iv in price_means.index],
                         rotation=45, ha='right', fontsize=8)
axes[0].axhline(0, color='gray', linestyle='--', alpha=0.5)
axes[0].set(xlabel='Nightly Price Range (USD)', ylabel='Mean Expectation Gap',
            title='Mean Expectation Gap vs. Price Range')

# Amenity count scatter
axes[1].scatter(listings_gap['amenity_count'], listings_gap['expectation_gap'],
                alpha=0.1, s=8, color='steelblue')
z = np.polyfit(listings_gap['amenity_count'].dropna(),
               listings_gap['expectation_gap'].dropna(), 1)
x_range = np.linspace(listings_gap['amenity_count'].min(), listings_gap['amenity_count'].max(), 100)
axes[1].plot(x_range, np.poly1d(z)(x_range), 'r--', linewidth=2, label='Trend')
axes[1].axhline(0, color='gray', linestyle='--', alpha=0.5)
axes[1].set(xlabel='Number of Amenities', ylabel='Expectation Gap',
            title='Expectation Gap vs. Amenity Count')
axes[1].legend()

plt.tight_layout()
plt.savefig('results/plot3_price_amenities.png', dpi=150, bbox_inches='tight')
plt.show()
```

The left plot groups listings into price ranges and shows the average Expectation Gap for each group. The right plot is a scatter plot showing every listing's amenity count against its Expectation Gap, with a trend line added.

**Price:** The average gap is negative across every price range, meaning guests are generally pleasantly surprised regardless of what they paid. There is no clear pattern,  cheaper and more expensive listings both have similar gap sizes. This suggests that how much a listing costs has nothing to do with how accurately the host describes it.

**Amenities:** The scatter plot shows a very slight downward trend, listings with more amenities have a marginally more negative gap (guests a bit more pleasantly surprised). ### Analysis

With the data explored, we now run three formal statistical tests to answer our research question. Each test is designed to address one specific part: whether superhost status matters, which listing features predict the gap, and whether a smaller gap actually leads to more bookings.


#### Analysis 1: Do Superhost Listings Have a Significantly Different Expectation Gap?

To test this, we use a **two-sample t-test**: a standard statistical test that checks whether two groups have different averages, and whether that difference is large enough to be unlikely due to chance. Here, the two groups are superhost listings and non-superhost listings, and we are comparing their average Expectation Gap scores.


```python
# Analysis 1: Independent samples t-test - Superhost vs Non-Superhost Expectation Gap

superhost_gap = listings_gap[listings_gap['host_is_superhost'] == True]['expectation_gap'].dropna()
non_superhost_gap = listings_gap[listings_gap['host_is_superhost'] == False]['expectation_gap'].dropna()

t_stat, p_val = stats.ttest_ind(superhost_gap, non_superhost_gap)

print("Superhost vs Non-Superhost - Expectation Gap:\n")
print(f"Superhost     mean gap: {superhost_gap.mean():.4f}  (n={len(superhost_gap):,})")
print(f"Non-Superhost mean gap: {non_superhost_gap.mean():.4f}  (n={len(non_superhost_gap):,})")
print(f"\nt-statistic: {t_stat:.4f}")
print(f"p-value:     {p_val:.2e}")
print(f"\n\u2192 {'Statistically significant' if p_val < 0.05 else 'Not significant'} at \u03b1 = 0.05")

# Visualization
fig, ax = plt.subplots(figsize=(7, 5))
bp = ax.boxplot([superhost_gap, non_superhost_gap],
                labels=['Superhost', 'Non-Superhost'], patch_artist=True,
                medianprops=dict(color='red', linewidth=2))
bp['boxes'][0].set_facecolor('lightblue')
bp['boxes'][1].set_facecolor('lightyellow')
ax.axhline(0, color='gray', linestyle='--', alpha=0.6, label='No gap (0)')
ax.set(ylabel='Expectation Gap',
       title=f'Expectation Gap by Superhost Status\n(t = {t_stat:.3f}, p = {p_val:.2e})')
ax.legend()
plt.tight_layout()
plt.savefig('results/analysis1_superhost_ttest.png', dpi=150, bbox_inches='tight')
plt.show()
```

The difference is statistically significant (t = −5.83, p < 0.001), meaning it is very unlikely to have occurred by chance. Superhost listings have an average gap of **−0.073**, compared to **−0.029** for non-superhost listings. In plain terms: superhost guests tend to be more pleasantly surprised by their stay relative to what the description promised.

Superhosts must maintain a rating of at least 4.8 and complete at least 10 stays per year to keep their badge, so they have a strong incentive to write honest, grounded descriptions rather than exaggerate. Non-superhosts have no such accountability mechanism, which may contribute to them being more likely to oversell.

Both groups still have negative average gaps overall, meaning the typical listing in both categories leaves guests slightly more impressed than the description implied. The difference is in how consistently this happens, and superhosts do it more reliably.


#### Analysis 2: Which Listing Features Best Predict the Expectation Gap?

To answer this, we run an **OLS linear regression**: a statistical model that estimates how much each feature contributes to the gap while holding all the other features constant. This directly tells us which predictors are statistically significant and in which direction they push the gap.

The predictors are: `desc_length` (description length), `amenity_count`, `price`, `host_tenure_years`, `calculated_host_listings_count`, `host_is_superhost`, `instant_bookable`, and `room_type` (categorical, reference = Entire home/apt).

```python
# Analysis 2: OLS Linear Regression - predicting the Expectation Gap

formula = (
    'expectation_gap ~ desc_length + amenity_count + price + '
    'host_tenure_years + calculated_host_listings_count + '
    'host_is_superhost + instant_bookable + '
    'C(room_type, Treatment("Entire home/apt"))'
)

reg_df = listings_gap.dropna(subset=[
    'expectation_gap', 'desc_length', 'amenity_count', 'price',
    'host_tenure_years', 'calculated_host_listings_count',
    'host_is_superhost', 'instant_bookable', 'room_type'
]).copy()
reg_df['host_is_superhost'] = reg_df['host_is_superhost'].astype(float)
reg_df['instant_bookable'] = reg_df['instant_bookable'].astype(float)

model = smf.ols(formula, data=reg_df).fit()

print(f"OLS Regression Results")
print(f"N = {int(model.nobs):,}   R\u00b2 = {model.rsquared:.4f}   Adj. R\u00b2 = {model.rsquared_adj:.4f}")
print(f"\nCoefficients (significant predictors at p < 0.05 marked with *):")
print(f"{'Predictor':<45} {'Coef':>10} {'p-value':>10}")
print("-" * 67)
for name in model.params.index:
    sig = '*' if model.pvalues[name] < 0.05 else ' '
    clean_name = (name.replace('C(room_type, Treatment("Entire home/apt"))[T.', 'Room: ')
                      .replace(']', ''))
    print(f"{clean_name:<45} {model.params[name]:>10.5f} {model.pvalues[name]:>10.4f} {sig}")

# Coefficient plot (significant predictors only)
sig_mask = model.pvalues < 0.05
sig_params = model.params[sig_mask].drop('Intercept', errors='ignore')
sig_ci = model.conf_int()[sig_mask].drop('Intercept', errors='ignore')

clean_labels = [
    (n.replace('C(room_type, Treatment("Entire home/apt"))[T.', 'Room: ').replace(']', ''))
    for n in sig_params.index
]

fig, ax = plt.subplots(figsize=(9, 5))
ax.barh(range(len(sig_params)), sig_params.values,
        xerr=[sig_params.values - sig_ci[0].values, sig_ci[1].values - sig_params.values],
        color='steelblue', alpha=0.8, capsize=4, ecolor='black')
ax.set_yticks(range(len(sig_params)))
ax.set_yticklabels(clean_labels, fontsize=9)
ax.axvline(0, color='black', linewidth=1)
ax.set(xlabel='Regression Coefficient (effect on Expectation Gap)',
       title='OLS Regression: Significant Predictors of Expectation Gap\n(error bars = 95% CI, p < 0.05)')
plt.tight_layout()
plt.savefig('results/analysis2_regression_coefs.png', dpi=150, bbox_inches='tight')
plt.show()
```

The model explains **14.3% of the variation** in Expectation Gap scores (R² = 0.143, n = 7,879). This means the listing features we measured account for about one-seventh of why gaps differ across listings; the rest comes from factors we do not observe, like the specific words a host uses, guest expectations going in, or the quality of the photos. For a metric built from noisy text data, 14% is a meaningful starting point.

**Description length** is the strongest predictor: every 100 extra characters in the description pushes the gap up by about +0.095. Hosts who write longer descriptions tend to oversell more. More words creates more opportunity to build up expectations that the actual experience cannot fully match.

**Superhost status** reduces the gap by 0.047 on average, even after controlling for everything else. This aligns with the t-test finding that superhost guests tend to be more pleasantly surprised.

**Instant bookable** listings have a slightly larger gap (+0.043), which may reflect that hosts who auto-accept all bookings are less attentive to managing what guests expect.

**Amenity count** has a small but significant negative effect (−0.00078 per amenity), meaning more amenities are associated with a slightly smaller gap, which supports the idea that concrete, specific features give hosts something real to describe.

**Host tenure** and **host listing count** both reduce the gap modestly, pointing to experience as a possible factor in writing more calibrated descriptions.

**Private room** listings have a slightly more negative gap than entire-home listings, meaning private room guests tend to be a bit more pleasantly surprised.

**Price is not significant** (p = 0.54): nightly price does not appear to be connected to how accurately a host describes their listing.


#### Analysis 3: Do Listings With Smaller Gaps Get More Bookings?

Our hypothesis predicts that listings where the description closely matches the experience will attract more guests over time, because accurate descriptions build trust. To test this, we split listings into two groups, those with a **small gap** (absolute gap below the median of 0.114) and those with a **large gap**, and compare their estimated number of booked nights per year using a t-test.

We use the absolute gap (|gap|) here rather than the signed gap because a large mismatch in either direction, whether the host oversold or dramatically undersold, could signal an unpredictable listing to potential guests.


```python
# Analysis 3: Booking demand - small-gap vs large-gap listings

median_abs_gap = listings_gap['expectation_gap'].abs().median()
listings_gap['gap_group'] = listings_gap['expectation_gap'].abs().apply(
    lambda x: 'Small Gap' if x < median_abs_gap else 'Large Gap'
)

small_gap_occ = listings_gap[listings_gap['gap_group'] == 'Small Gap']['estimated_occupancy_l365d'].dropna()
large_gap_occ = listings_gap[listings_gap['gap_group'] == 'Large Gap']['estimated_occupancy_l365d'].dropna()

t_stat2, p_val2 = stats.ttest_ind(small_gap_occ, large_gap_occ)
r_occ = listings_gap[['expectation_gap', 'estimated_occupancy_l365d']].dropna().corr().iloc[0, 1]

print(f"Booking Demand: Small Gap vs Large Gap Listings:\n")
print(f"Median |Expectation Gap|: {median_abs_gap:.4f}")
print(f"\nSmall gap (|gap| < {median_abs_gap:.3f}): mean {small_gap_occ.mean():.1f} nights/year  (n={len(small_gap_occ):,})")
print(f"Large gap (|gap| \u2265 {median_abs_gap:.3f}): mean {large_gap_occ.mean():.1f} nights/year  (n={len(large_gap_occ):,})")
print(f"\nt-statistic: {t_stat2:.4f}")
print(f"p-value:     {p_val2:.2e}")
print(f"Pearson r (gap vs occupancy): {r_occ:.4f}")
print(f"\n\u2192 {'Statistically significant' if p_val2 < 0.05 else 'Not significant'} at \u03b1 = 0.05")

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Boxplot
bp3 = axes[0].boxplot([small_gap_occ, large_gap_occ],
                      labels=['Small Gap', 'Large Gap'], patch_artist=True,
                      medianprops=dict(color='red', linewidth=2))
bp3['boxes'][0].set_facecolor('lightgreen')
bp3['boxes'][1].set_facecolor('lightsalmon')
axes[0].set(ylabel='Estimated Occupancy (nights/year)',
            title=f'Occupancy by Gap Group\n(t = {t_stat2:.3f}, p = {p_val2:.2e})')

# Scatter: |gap| vs occupancy
plot_occ = listings_gap[['expectation_gap', 'estimated_occupancy_l365d']].dropna()
axes[1].scatter(plot_occ['expectation_gap'].abs(), plot_occ['estimated_occupancy_l365d'],
                alpha=0.08, s=8, color='steelblue')
axes[1].set(xlabel='|Expectation Gap|', ylabel='Estimated Occupancy (nights/year)',
            title=f'Absolute Gap vs. Booking Demand\n(Pearson r = {r_occ:.3f})')

plt.tight_layout()
plt.savefig('results/analysis3_booking_demand.png', dpi=150, bbox_inches='tight')
plt.show()
```

The result is statistically significant (t = 5.63, p < 0.001). Listings with a small expectation gap average **137 booked nights per year**, compared to **126 nights** for large-gap listings. That is about 11 extra nights per year, or roughly 8.7% more bookings.

At San Diego's median nightly price of $175, those 11 nights translate to approximately **$1,925 in extra annual revenue** per listing. The effect is modest but consistent.

One additional note: when we compute the regular (signed) correlation between the gap and occupancy, it appears weak (r = 0.04). This is because positive and negative gaps cancel each other out in a standard correlation. Using the absolute gap, which treats overselling and underselling equally as forms of mismatch, better reflects what we are trying to measure and reveals a clearer difference between the two groups.

Overall, this finding is consistent with our hypothesis: hosts who set accurate expectations may attract more bookings over the course of a year.

Variables that we initially thought were important, like amenity count, showed little to no meaningful relationship with the expectation gap. Our analysis sugguests that structural aspects of the listings do not strongly influence guest dissatisfaction. Instead, the gap between expectation and experience appears to be shaped by how hosts market their listings, rather than the physical features of the property itself. This suggests that expectation disconfirmation is less about objective listing quality, and more about the presentation used in the description of Airbnb listings.

5. There are several limitations to be considered when interpreting our results. The dataset only represents a snapshot of Airbnb listings in San Diego from September 2025, meaning that our findings may not generalize other cities or time periods where tourism patterns, housing markets, and guest expectations may differ. Additionally, the Expectation Gap metric relies on VADER sentiment analysis, which has known limitations when applied to short texts, subtle phrasing, or sarcastic language. Airbnb reviews are also strongly skewed toward positive sentiment, which compresses variation in review scores and may obscure more nuanced forms of dissatisfaction. The regression model explains only about 14% of the variation in expectation gaps, suggesting that many relevant factors such as listing photos, the specific wording used in descriptions, or guest expectations before booking are not captured in our dataset. Due to these limitations, the results should be interpreted as an initial exploration rather than a complete explanation of why expectation gaps occur.
