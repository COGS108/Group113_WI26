## Background and Prior Work

<font color="red">Instructions: REPLACE the contents of this cell with your work, including any updates to recover points lost in your proposal feedback</font>

Online peer-to-peer (P2P) marketplaces rely heavily on trust between users, and platforms like Airbnb are no exception. Because guests cannot physically visit a property before booking, they depend largely on host-provided listing descriptions, photos, and amenities to form expectations about their stay. Expectation‚Äìconfirmation theory (ECT), developed by Oliver and widely applied in hospitality contexts, models satisfaction as this comparison between prior expectations and perceived performance; negative disconfirmation (worse than expected) leads to dissatisfaction, while confirmation or positive disconfirmation leads to satisfaction.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) When guests' experiences deviate significantly from their expectations, the resulting expectation gap can lead to lower ratings, negative reviews, reduced future demand, and erosion of platform trust.<br><br>
‚Äã
Text‚Äëmining work on Airbnb reviews shows what guests actually care about. Cheng and Jin analyze a large corpus of Airbnb review comments using topic modeling and sentiment analysis, finding that experiences are dominated by location, amenities, and host interaction, with a strong overall positivity bias in reviews.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Other work confirms that Airbnb reviews are overwhelmingly positive and that criticism is often muted, which makes star ratings alone a poor indicator of underlying problems.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)<br>

Expectation disconfirmation has been explicitly studied in the Airbnb context: Zhang et al. show that mismatches between expectations and experience, and how consistent reviews are, meaningfully affect guest attitudes and intention to recommend listings.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) Most directly related to this project, Cavique et al. measure semantic alignment between Airbnb host descriptions and guest reviews using word embeddings and clustering. They introduce a concept coverage ratio (how many review ‚Äúconcepts‚Äù are also present in the description) and find that higher coverage is generally associated with more positive review sentiment.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)<br><br>
‚Äã
Building on this work, the proposed project defines a new expectation disconfirmation score that combines global semantic similarity between descriptions and reviews with host‚Äìguest sentiment differences (capturing overpromising vs. underpromising). It then uses supervised models to predict this score from pre-booking information (descriptions, amenities, host and listing features). In a diverse tourism market like San Diego, where listings range from luxury beachfront villas in La Jolla to urban studios in Little Italy, the factors contributing to expectation gaps are multifaceted. By quantifying which listing characteristics predict smaller expectation gaps, this project aims to provide actionable insights for hosts to improve transparency and for platforms to better calibrate guest expectations.

Our data comes from [Inside Airbnb](http://insideairbnb.com/get-the-data/), which is an independent open project that scrapes publicly available Airbnb data for cities around the world. The data is free to download and no application or permission needed. We will be using **San Diego** data from **September 2025**.

**Dataset 1: `listings.csv.gz`**

This is the main file. It has 13,162 San Diego listings with 79 columns. Below are the relevant columns we have selected:

| Column | Description | Purpose |
|---|---|---|
| `description` | The host's written description of their place | We run VADER sentiment analysis on this to get the host-side sentiment score |
| `host_is_superhost` | Superhost badge (t/f) | Predictor |
| `price` | Nightly price (stored as a string with `$`, needs cleaning) | Predictor |
| `neighbourhood_cleansed` | Neighborhood name | Predictor |
| `room_type` | Entire home, private room, shared room, or hotel room | Predictor |
| `amenities` | JSON list of amenities | We count these per listing and use the count as a predictor |
| `review_scores_rating` | Overall guest rating (0‚Äì5) | Used to validate our gap metric |
| `review_scores_accuracy` | Accuracy sub-score from guests | This is the closest existing metric to what our gap tries to capture |
| `number_of_reviews` | Total review count | We use this to filter to only keep listings with 5+ reviews |
| `reviews_per_month` | Avg reviews per month | Substitute for how often the listing gets booked |

The key variables we use are `description` (the host's written listing text, which we sentiment-score), `host_is_superhost`,`price`, `neighbourhood_cleansed`, `room_type`, and `amenities` as predictor variables, along with `review_scores_rating` and `review_scores_accuracy` for validation. We will use `reviews_per_month` as a substitute for booking frequency. After we filter out listings that are missing a `description` or have fewer than 5 reviews in `number_of_reviews`, we're left with **9,091 listings** as usable observations. The superhost split is roughly 60/40, which gives us a solid basis for comparison.

**Dataset 2: `reviews.csv.gz`**

This dataset is also available from Inside Airbnb at the same URL and download page as the listings data, with no access restrictions. The file contains 896,198 individual reviews across 10,962 listings. Below are the relevant columns we have selected:

| Column | Description | Purpose |
|---|---|---|
| `listing_id` | Links the review to a listing | Join key to connect reviews back to listings |
| `comments` | The actual review text | We run VADER sentiment analysis on this for the guest-side sentiment |

The two variables we use are `listing_id`, which links each review back to a listing so we can join the two datasets, and `comments`, which contains the full review text that we run VADER sentiment analysis on to compute the guest-side sentiment score of the expectation gap.

Other available datasets from Inside Airbnb (San Diego, September 2025) not used in this project:

- **`calendar.csv.gz`**: daily availability and pricing for each listing over 365 days; excluded because "unavailable" dates don't reliably indicate bookings (hosts block dates manually), making it a noisy proxy that correlates almost zero with `reviews_per_month`.
- **`neighbourhoods.csv`**: a reference list of 108 neighborhood names; excluded because this info is already in the `neighbourhood_cleansed` column of the listings file.

### Real dataset vs. Ideal dataset

Our real datasets closely match the ideal. The main differences are:

- **No direct booking data.** Airbnb does not publicly share booking counts, so we use `reviews_per_month` as a substitute for booking frequency, which is standard in Airbnb research but is an approximation.
- **Single snapshot in time.** The data represents a single scrape (September 2025) rather than a longitudinal panel, so we cannot track how the Expectation Gap changes over time for a given listing.
- **One city only.** Our analysis is limited to San Diego; findings may not generalize to other markets with different guest demographics or listing characteristics.
- **Noisy reviews.** Some reviews are super short ("Great stay!") or written in other languages, which makes individual sentiment scores less reliable. But since we're averaging across multiple reviews per listing and requiring at least 5, aggregating across multiple reviews per listing mitigates this.

Despite these limitations, the dataset is large, detailed, and contains all the core variables needed to for our analysis.
