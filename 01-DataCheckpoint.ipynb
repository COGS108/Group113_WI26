## Research Question

<font color="red">Instructions: REPLACE the contents of this cell with your work, including any updates to recover points lost in your proposal feedback</font>



**How can expectation disconfirmation between San Diego Airbnb host descriptions and guest experiences be quantified from text and reviews?**

- What factors from the listings explain why some listings have lower expectation gap compared to others? 
- Compute expectation disconfirmation by using the hosting description and guests reviews
- Examine whether listing features like superhost status, price, neighborhood, room type, amenities, description language from host listing to predict the expectation gap
- How does the expectation disconfirmation relate to outcomes like review scores and booking frequency? 

According to Oliver (1977), **disconfirmation** is defined as the extent to which a consumer's initial expectations about a product's performance are met. Specifically, negative disconfirmation occurs when a product performs worse than expected, confirmation happens when it performs exactly as expected, and positive disconfirmation occurs when it performs better than anticipated.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1)

For our project, we measure this disconfirmation using a metric called the **Expectation Gap**, defined as follows:

> **Expectation Gap = VADER(description) − mean(VADER(reviews))**

where VADER produces a compound sentiment score from −1 (most negative) to +1 (most positive). A gap near 0 indicates the description sentiment matched review sentiment; a large positive gap (> 0) means the host oversold the listing (description more positive than guest experience); a negative gap means the host undersold it (guests were pleasantly surprised).

Secondary questions:

- Which listing features best predict a smaller Expectation Gap (absolute gap closer to 0)?
- How does the Expectation Gap relate to overall review scores (review_scores_rating, review_scores_accuracy) and estimated booking frequency (estimated_occupancy_l365d)?

## Background and Prior Work

<font color="red">Instructions: REPLACE the contents of this cell with your work, including any updates to recover points lost in your proposal feedback</font>

Online peer-to-peer (P2P) marketplaces rely heavily on trust between users, and platforms like Airbnb are no exception. Because guests cannot physically visit a property before booking, they depend largely on host-provided listing descriptions, photos, and amenities to form expectations about their stay. Expectation–confirmation theory (ECT), developed by Oliver and widely applied in hospitality contexts, models satisfaction as this comparison between prior expectations and perceived performance; negative disconfirmation (worse than expected) leads to dissatisfaction, while confirmation or positive disconfirmation leads to satisfaction.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) <a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) When guests' experiences deviate significantly from their expectations, the resulting expectation gap can lead to lower ratings, negative reviews, reduced future demand, and erosion of platform trust.<br><br>
​
Text‑mining work on Airbnb reviews shows what guests actually care about. Cheng and Jin analyze a large corpus of Airbnb review comments using topic modeling and sentiment analysis, finding that experiences are dominated by location, amenities, and host interaction, with a strong overall positivity bias in reviews.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) Other work confirms that Airbnb reviews are overwhelmingly positive and that criticism is often muted, which makes star ratings alone a poor indicator of underlying problems.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4)<br>

Expectation disconfirmation has been explicitly studied in the Airbnb context: Zhang et al. show that mismatches between expectations and experience, and how consistent reviews are, meaningfully affect guest attitudes and intention to recommend listings.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) Most directly related to this project, Cavique et al. measure semantic alignment between Airbnb host descriptions and guest reviews using word embeddings and clustering. They introduce a concept coverage ratio (how many review “concepts” are also present in the description) and find that higher coverage is generally associated with more positive review sentiment.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2)<br><br>
​
Building on this work, the proposed project defines a new expectation disconfirmation score that combines global semantic similarity between descriptions and reviews with host–guest sentiment differences (capturing overpromising vs. underpromising). It then uses supervised models to predict this score from pre-booking information (descriptions, amenities, host and listing features). In a diverse tourism market like San Diego, where listings range from luxury beachfront villas in La Jolla to urban studios in Little Italy, the factors contributing to expectation gaps are multifaceted. By quantifying which listing characteristics predict smaller expectation gaps, this project aims to provide actionable insights for hosts to improve transparency and for platforms to better calibrate guest expectations.

**Dataset concerns and shortcomings:**

There are some concerns associated with this dataset based on the disclaimers and data assumptons provided by the scraping website.

- **Single snapshot:** 
The dataset is composed of listings that were available at the time of scraping, which was done September 25, 2025. This means this dataset does not capture any listings deleted before or after this period. As a result, the sample may not fully represent the dynamic nature of the Airbnb market, and ay conclusions drawn apply only to listings active on September 25, 2025.

- **No actual booking data:** Airbnb does not publish booking counts. Both `reviews_per_month` and `estimated_occupancy_l365d` are derived from review counts, not direct booking records. Standard in Airbnb research but are approximations.

- **Availability ≠ bookings:** 
The Airbnb calendar for a listing does not differentiate between a booked night VS an unavailable night. All of these nights have been grouped into "unavailable". This introduces measurement bias, as highly popular listings are more likely to appear unavailable dure to bookings rather than host restrictions. Consequently, the data may systematically understate availability for successful listings and could misrepresent listing popularity if availability is used as a substitute for demand. This is why we are using occupancy estimates instead.

- **Neighborhood geocoding:** 
Neighbourhood names for each listing are compiled by comparing the listing's geographic coordinates with a city's definition of neighbourhoods. Listings located near neighborhood borders may be misclassified if small coordinate errors or boundary discrepancies exist. Also, official neighborhood boundaries may not reflect how guests or hosts socially perceive or describe an area. Airbnb also anonymizes coordinates by 0~450 feet (150m).

- **Price outliers:** 
After exploring the data, we found out that out of all the San Diego Airbnb listings, 28 of the listings cost over $2000 a night, up to the $50k range. We did flag it out as an outlier, but kept for sensitivity analysis.


- **Description truncation:** 
It looks like the descriptions are capped at ~1,000 characters in our scraped data. This means any description exceeding 1,000 characters can not be reflected in our analysis.

### Data set 1: Data Cleaning, Filtering, Outlier Detection

```python
# Data set 1: Data Cleaning, Filtering, Outlier Detection

# ---- Clean price: '$175.00' → 175.0 ----
listings['price'] = (
    listings['price']
    .str.replace('$', '', regex=False)
    .str.replace(',', '', regex=False)
    .astype(float)
)
print('Price after cleaning:')
print(listings['price'].describe().round(2))

# ---- Clean boolean columns: 't'/'f' → True/False ----
for col in ['host_is_superhost', 'host_identity_verified', 'instant_bookable']:
    listings[col] = listings[col].map({'t': True, 'f': False})
print(f'\nBoolean columns cleaned: host_is_superhost, host_identity_verified, instant_bookable')

# ---- Clean host_response_rate and host_acceptance_rate: '95%' → 0.95 ----
for col in ['host_response_rate', 'host_acceptance_rate']:
    listings[col] = (
        listings[col]
        .str.replace('%', '', regex=False)
        .astype(float) / 100
    )
print(f'Percentage columns cleaned: host_response_rate (mean={listings["host_response_rate"].mean():.3f}), '
      f'host_acceptance_rate (mean={listings["host_acceptance_rate"].mean():.3f})')

# ---- Parse bathrooms_text ----
# Extracts TWO things:
#   bathrooms  — numeric count (e.g., 2.5; "Half-bath" → 0.5)
#   bath_type  — 'private', 'shared', or 'unspecified' (host didn't say either way)
def parse_bathrooms(text):
    if pd.isna(text):
        return np.nan
    text_lower = text.lower()
    if 'half-bath' in text_lower or 'half bath' in text_lower:
        return 0.5
    match = re.search(r'(\d+\.?\d*)', text)
    return float(match.group(1)) if match else np.nan

def parse_bath_type(text):
    # 'unspecified' means text just says e.g. '2 baths' — for entire homes
    #this is likely private, but the host didn't explicitly say.
    if pd.isna(text):
        return np.nan
    text_lower = text.lower()
    if 'private' in text_lower:
        return 'private'
    elif 'shared' in text_lower:
        return 'shared'
    return 'unspecified'

listings['bathrooms'] = listings['bathrooms_text'].apply(parse_bathrooms)
listings['bath_type'] = listings['bathrooms_text'].apply(parse_bath_type)
print(f'\nBathrooms parsed — bath_type distribution:')
print(listings['bath_type'].value_counts(dropna=False).to_string())

# ---- Parse host_since → host_tenure_years ----
listings['host_since'] = pd.to_datetime(listings['host_since'])
scrape_date = pd.Timestamp('2025-09-25')
listings['host_tenure_years'] = ((scrape_date - listings['host_since']).dt.days / 365.25).round(1)
print(f'\nHost tenure: mean={listings["host_tenure_years"].mean():.1f} yrs, '
      f'median={listings["host_tenure_years"].median():.1f} yrs')

# ---- Extract amenity count from JSON list ----
def count_amenities(amenity_str):
    if pd.isna(amenity_str):
        return 0
    try:
        return len(json.loads(amenity_str))
    except (json.JSONDecodeError, TypeError):
        return amenity_str.count(',') + 1

listings['amenity_count'] = listings['amenities'].apply(count_amenities)
print(f'Amenity count: mean={listings["amenity_count"].mean():.1f}, '
      f'median={listings["amenity_count"].median():.0f}')

# ---- Strip HTML from description, compute length ----
listings['description_clean'] = (
    listings['description']
    .fillna('')
    .str.replace(r'<br\s*/?>', ' ', regex=True)
    .str.replace(r'<[^>]+>', '', regex=True)
    .str.strip()
)
listings['desc_length'] = listings['description_clean'].str.len()
print(f'Description length (HTML stripped): mean={listings["desc_length"].mean():.0f} chars, '
      f'median={listings["desc_length"].median():.0f} chars')

# ---- Filter: must have description (≥10 chars) AND ≥5 reviews ----
print('\n'+ 50*'-')
print('FILTERING')
print(50*'-')
no_desc = listings['description_clean'].str.len() < 10
few_reviews = listings['number_of_reviews'] < 5
print(f'Total listings:             {len(listings):,}')
print(f'No/short description:       {no_desc.sum():,} ({no_desc.mean():.1%})')
print(f'Fewer than 5 reviews:       {few_reviews.sum():,} ({few_reviews.mean():.1%})')
listings_clean = listings[~no_desc & ~few_reviews].copy()
print(f'After filter:               {len(listings_clean):,}')

# Drop rows still missing price or superhost (can't use as predictors)
# host_response_time/rate, bedrooms, beds, bathrooms NaN → keep (handle in modeling)
before = len(listings_clean)
listings_clean = listings_clean.dropna(subset=['price', 'host_is_superhost'])
print(f'Dropped {before - len(listings_clean)} rows missing price or superhost → {len(listings_clean):,} remaining')

# ---- Outlier detection: price ----
print('\n'+ 50*'-')
print('OUTLIER DETECTION')
print(50*'-')
for t in [1000, 2000, 5000, 10000]:
    print(f'  Price > ${t:,}/night: {(listings_clean["price"] > t).sum()} listings')
listings_clean['price_outlier'] = listings_clean['price'] > 2000
print(f'Flagged as price outlier (>$2k): {listings_clean["price_outlier"].sum()} (kept in data for sensitivity analysis)')

# Outlier visualizations
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].hist(listings_clean['price'].clip(upper=1500), bins=50, edgecolor='black', alpha=0.7)
axes[0].set_title('Price Distribution (capped at $1,500)')
axes[0].set_xlabel('Price ($/night)')
axes[0].set_ylabel('Count')
axes[1].hist(listings_clean['amenity_count'], bins=40, edgecolor='black', alpha=0.7)
axes[1].set_title('Amenity Count Distribution')
axes[1].set_xlabel('Number of amenities')
axes[2].hist(listings_clean['estimated_occupancy_l365d'], bins=40, edgecolor='black', alpha=0.7)
axes[2].set_title('Estimated Occupancy (last 365 days)')
axes[2].set_xlabel('Nights booked')
plt.tight_layout()
plt.show()
```
