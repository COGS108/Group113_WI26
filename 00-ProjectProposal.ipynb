## Background and Prior Work


Online peer-to-peer (P2P) marketplaces rely heavily on trust between users, and platforms like Airbnb are no exception. Because guests cannot physically visit a property before booking, they depend largely on host-provided listing descriptions, photos, and amenities to form expectations about their stay. When their stay deviates from their expectations, it results in an expectation gap. A large gap may lead to lower ratings and negative sentiment in reviews, which can negatively impact host ratings, future demand, and platform trust overall.


In a diverse tourism market like San Diego, the factors that contribute to this gap are multifaceted. Listings range from luxury beachfront villas in La Jolla to urban studios in Little Italy, each carrying different “baseline” expectations. This project investigates which listing factors predict a smaller expectation gap for San Diego Airbnb listings. By quantifying the alignment between host descriptions and guest reviews, we aim to identify the specific factors that lead to a more accurate expectation setting. These insights provide a roadmap for hosts to improve transparency and for the platform to better calibrate guest expectations.

Text‑mining work on Airbnb reviews shows what guests actually care about. Cheng and Jin analyze a large corpus of Airbnb review comments using topic modeling and sentiment analysis, finding that experiences are dominated by location, amenities, and host interaction, with a strong overall positivity bias in reviews. Other work confirms that Airbnb reviews are overwhelmingly positive and that criticism is often muted, which makes star ratings alone a poor indicator of underlying problems. (#cite_note-1) (#cite_note-2)


Expectation disconfirmation has been explicitly studied in the Airbnb context: Zhang et al. show that mismatches between expectations and experience, and how consistent reviews are, meaningfully affect guest attitudes and intention to recommend listings. Most directly related to this project, Cavique et al. measure semantic alignment between Airbnb host descriptions and guest reviews using word embeddings and clustering. They introduce a concept coverage ratio (how many review “concepts” are also present in the description) and find that higher coverage is generally associated with more positive review sentiment. (#cite_note-3) (#cite_note-4)

## Data

### Ideal Dataset

For our question: What factors predict a smaller expectation score between how Airbnb hosts describe their listings and what guests actually report in their reviews for San Diego listings? The ideal dataset would have the following: 

- The full text of each host's listing description, for sentiment analysis
- The full text of every guest review for each listing, to obtain the guest-side sentiment
- Listing-level features such as superhost status, price, neighborhood, room type, and amenities, since these are the predictors we plan to test
- Overall Rating metric to cross check our computed expectation score against what guest rated
- Some measure of how often a listing gets booked, to see if the gap affects demand

For the dataset to be around 5,000 to 10,000 listings, each with at least 5 reviews so the per-listing sentiment averages aren't thrown off by one or two outlier reviews. That means we'd also need hundreds of thousands of individual review texts to cover all those listings.The idea data would be cleanly organized in tables for example a listings table (one row per listing with all the attributes) and a reviews table (one row per review, linked back to the listing by ID).

### Real Datasets

Our actual data comes from [Inside Airbnb](http://insideairbnb.com/get-the-data/), which is an independent open project that scrapes publicly available Airbnb data for cities around the world. The data is free to download and no application or permission needed. We will be using **San Diego** data from **September 2025**.

**Dataset 1: `listings.csv.gz`**

This is the main file. It has 13,162 San Diego listings with 79 columns. Below are the relevant columns we have selected:

| Column | Description | Purpose |
|---|---|---|
| `description` | The host's written description of their place | We run VADER sentiment analysis on this to get the host-side score |
| `host_is_superhost` | Superhost badge (t/f) | Predictor |
| `price` | Nightly price (stored as a string with `$`, needs cleaning) | Predictor |
| `neighbourhood_cleansed` | Neighborhood name | Predictor |
| `room_type` | Entire home, private room, shared room, or hotel room | Predictor |
| `amenities` | JSON list of amenities | We count these per listing and use the count as a predictor |
| `review_scores_rating` | Overall guest rating (0–5) | Used to validate our gap metric |
| `review_scores_accuracy` | Accuracy sub-score from guests | This is the closest existing metric to what our gap tries to capture |
| `number_of_reviews` | Total review count | We use this to filter to only keep listings with 5+ reviews |
| `reviews_per_month` | Avg reviews per month | Substitute for how often the listing gets booked |

The key variables we use are `description` (the host's written listing text, which we sentiment-score), `host_is_superhost`,`price`, `neighbourhood_cleansed`, `room_type`, and `amenities` as predictor variables, along with `review_scores_rating` and `review_scores_accuracy` for validation. We will use `reviews_per_month` as a substitute for booking frequency. After we filter out listings that are missing a `description` or have fewer than 5 reviews in `number_of_reviews`, we're left with **9,091 listings** as usable observations. The superhost split is roughly 60/40, which gives us a solid basis for comparison.

**Dataset 2: `reviews.csv.gz`**

This dataset is also available from Inside Airbnb at the same URL and download page as the listings data, with no access restrictions. The file contains 896,198 individual reviews across 10,962 listings. Below are the relevant columns we have selected:

| Column | Description | Purpose |
|---|---|---|
| `listing_id` | Links the review to a listing | Join key to connect reviews back to listings |
| `comments` | The actual review text | We run VADER sentiment analysis on this for the guest-side score |

The two variables we use are `listing_id`, which links each review back to a listing so we can join the two datasets, and `comments`, which contains the full review text that we run VADER sentiment analysis on to compute the guest-side score of the expectation score.

Other available datasets from Inside Airbnb (San Diego, September 2025) not used in this project:

- **`calendar.csv.gz`**: daily availability and pricing for each listing over 365 days; excluded because "unavailable" dates don't reliably indicate bookings (hosts block dates manually), making it a noisy proxy that correlates almost zero with `reviews_per_month`.
- **`neighbourhoods.csv`**: a reference list of 108 neighborhood names; excluded because this info is already in the `neighbourhood_cleansed` column of the listings file.

### Real dataset vs. Ideal dataset

Our real datasets closely match the ideal. The main differences are:

- **No direct booking data.** Airbnb does not publicly share booking counts, so we use `reviews_per_month` as a substitute for booking frequency, which is standard in Airbnb research but is an approximation.
- **Single snapshot in time.** The data represents a single scrape (September 2025) rather than a longitudinal panel, so we cannot track how the Expectation Gap changes over time for a given listing.
- **One city only.** Our analysis is limited to San Diego; findings may not generalize to other markets with different guest demographics or listing characteristics.
- **Noisy reviews.** Some reviews are super short ("Great stay!") or written in other languages, which makes individual sentiment scores less reliable. But since we're averaging across multiple reviews per listing and requiring at least 5, aggregating across multiple reviews per listing mitigates this.

Despite these limitations, the dataset is large, detailed, and contains all the core variables needed to for our analysis.
